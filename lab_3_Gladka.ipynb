код для вимірювання часу та побудови графіків та бульбашкове сортування

```python
import time
import random
import matplotlib.pyplot as plt

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 
         10000, 20000, 50000, 100000]

t_insert, t_bubble = [], []

for n in sizes:
    arr = [random.randint(0, 10000) for _ in range(n)]
    
    start = time.time()
    insertion_sort(arr.copy())
    t_insert.append(time.time() - start)
    
    start = time.time()
    bubble_sort(arr.copy())
    t_bubble.append(time.time() - start)

plt.plot(sizes, t_insert, label="Insertion Sort", marker="o")
plt.plot(sizes, t_bubble, label="Bubble Sort", marker="o")
plt.xlabel("Розмір масиву n")
plt.ylabel("Час виконання (секунди)")
plt.title("Порівняння часу сортування")
plt.legend()
plt.grid(True)
plt.show()
```

Контрольні питання

**1. Що таке алгоритм пошуку і чому він важливий у контексті комп'ютерних наук?**  
Алгоритм пошуку — це метод знаходження потрібного елемента у структурі даних; він важливий, бо забезпечує швидкий доступ до інформації та ефективність програм.

**2. Які основні критерії оцінки ефективності алгоритмів пошуку?**  
Часова складність, кількість порівнянь, використання пам’яті, здатність працювати з відсортованими чи невідсортованими даними.

**3. Що таке лінійний пошук, і як він працює?**  
[Лінійний пошук](ca://s?q=Лінійний_пошук) послідовно перевіряє кожен елемент списку, доки не знайде потрібний або не дійде до кінця.

**4. Які умови повинні бути виконані для успішного застосування бінарного пошуку?**  
Список має бути **відсортованим**, і доступ до елементів має бути можливим за індексом.

**5. Які переваги та недоліки використання бінарного пошуку порівняно з іншими алгоритмами пошуку?**  
Переваги: швидкість \(O(\log n)\), менше порівнянь. Недоліки: працює лише з відсортованими даними, потребує випадкового доступу.

**6. Що таке тернарний пошук, і в чому його відмінність від бінарного пошуку?**  
[Тернарний пошук](ca://s?q=Тернарний_пошук) ділить список на три частини й порівнює з двома середніми елементами; відмінність у тому, що він робить більше порівнянь на кожному кроці, але має таку ж асимптотику \(O(\log n)\).
