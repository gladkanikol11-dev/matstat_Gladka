```python
def merge_sort(arr):
   
    if len(arr) <= 1:
        return arr
  
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0
  
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    
   
    result.extend(left[i:])
    result.extend(right[j:])
    return result


data = [38, 27, 43, 3, 9, 82, 10]
print("До:", data)
print("Після:", merge_sort(data))
```

Пояснення
Алгоритм рекурсивно ділить масив навпіл.

Кожна половина сортується окремо.

Потім дві відсортовані частини зливаються в один масив.

Асимптотична складність: Θ(n log n) (як ми вже розібрали за головною теоремою рекурсії).
