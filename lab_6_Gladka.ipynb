1. Приклад 3 з офіційної документації модуля Queue  
У Python модуль queue використовується для створення черг, які підтримують операції insert (put) та delete (get) з гарантією потокобезпечності. Приклад 3 демонструє роботу багатопотокових черг.

```python
from queue import Queue

def print_n(q, n):
    """
    Друкує перші n елементів черги включно.
    q - об'єкт Queue
    n - кількість елементів для друку
    """
    items = list(q.queue)  # доступ до внутрішнього списку черги
    for i in range(min(n, len(items))):
        print(items[i])
```
