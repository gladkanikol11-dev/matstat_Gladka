```python
def z_func(s):
    n = len(s)
    z = [0]*n
    l, r = 0, 0
    for i in range(1, n):
        if i <= r:
            z[i] = min(r-i+1, z[i-l])
        while i+z[i] < n and s[z[i]] == s[i+z[i]]:
            z[i] += 1
        if i+z[i]-1 > r:
            l, r = i, i+z[i]-1
    return z

def compress_with_z(s):
    z_vec = z_func(s)
    for i in range(1, len(s)):
        if (i + z_vec[i] == len(s)) and (len(s) % i == 0):
            return s[:i]
    return s

# приклад
s = "abcabcabcabc"
print("Z-вектор:", z_func(s))
print("Стиснене представлення:", compress_with_z(s))
```
