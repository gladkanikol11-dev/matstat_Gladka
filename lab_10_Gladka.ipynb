Побудувати дерево кодів Гафмена вручну

```python
import heapq
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter

class Node:
    def __init__(self, char=None, freq=0, left=None, right=None):
        self.char = char
        self.freq = freq
        self.left = left
        self.right = right
    def __lt__(self, other):
        return self.freq < other.freq
def build_huffman_tree(text):
    freq = Counter(text)
    heap = [Node(char, f) for char, f in freq.items()]
    heapq.heapify(heap)
    while len(heap) > 1:
        left = heapq.heappop(heap)
        right = heapq.heappop(heap)
        merged = Node(freq=left.freq+right.freq, left=left, right=right)
        heapq.heappush(heap, merged)
    return heap[0]
def build_codes(node, prefix="", code_map={}):
    if node.char is not None:
        code_map[node.char] = prefix
    else:
        build_codes(node.left, prefix+"0", code_map)
        build_codes(node.right, prefix+"1", code_map)
    return code_map
text = "ABABCAADEEGGFGFABBABAABCBAADEEGFGF"
root = build_huffman_tree(text)
codes = build_codes(root)
print("Коди символів:", codes)
def visualize_tree(node, graph=None, pos=None, parent=None, edge_label=""):
    if graph is None:
        graph = nx.DiGraph()
        pos = {}
    label = f"{node.char if node.char else ''}:{node.freq}"
    graph.add_node(label)
    pos[label] = (len(pos), -node.freq)
    if parent:
        graph.add_edge(parent, label, label=edge_label)
    if node.left:
        visualize_tree(node.left, graph, pos, label, "0")
    if node.right:
        visualize_tree(node.right, graph, pos, label, "1")
    return graph, pos
G, pos = visualize_tree(root)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue")
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u,v):d['label'] for u,v,d in G.edges(data=True)})
plt.show()
```

```python
def get_chars_and_freq(message):
    freq = Counter(message)
    chars = list(freq.keys())
    counts = list(freq.values())
    return chars, counts

chars, freq = get_chars_and_freq(text)
print("Символи:", chars)
print("Частоти:", freq)
```

```python
def huffman_decode(encoded, root):
    decoded = []
    node = root
    for bit in encoded:
        node = node.left if bit == "0" else node.right
        if node.char is not None:
            decoded.append(node.char)
            node = root
    return "".join(decoded)

# Приклад: закодувати і декодувати
encoded = "".join(codes[ch] for ch in text)
decoded = huffman_decode(encoded, root)
print("Закодоване повідомлення:", encoded)
print("Декодоване повідомлення:", decoded)
```

Кодування Гафмена будує оптимальне префіксне дерево для символів.

Складність побудови дерева: 
𝑂(𝑛log𝑛), де 𝑛 — кількість різних символів.
Складність кодування/декодування: 
𝑂(𝐿), де 𝐿 — довжина повідомлення.
Ефект: скорочення довжини повідомлення на 15–30% у порівнянні з фіксованим кодуванням.
