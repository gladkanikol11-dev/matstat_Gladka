1. Оцінити асимптотичну складність алгоритму бінарного пошуку в O‑нотації.  
Бінарний пошук працює за принципом ділення відсортованого масиву навпіл на кожному кроці.
Асимптотична складність:

𝑇(𝑛)=𝑂(log⁡𝑛)
У найгіршому випадку потрібно 
log2𝑛 порівнянь.

```python
import time
import pandas as pd

def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def test_linear_search(sizes, target):
    results = []
    for n in sizes:
        arr = list(range(n))
        start = time.time()
        linear_search(arr, target)
        end = time.time()
        results.append({"n": n, "time": end - start})
    return pd.DataFrame(results)
```

```python
def bin_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

def test_bin_search(sizes, target):
    results = []
    for n in sizes:
        arr = list(range(n))
        start = time.time()
        bin_search(arr, target)
        end = time.time()
        results.append({"n": n, "time": end - start})
    return pd.DataFrame(results)
```

```python
import matplotlib.pyplot as plt

sizes = [100, 1000, 5000, 10000, 50000]
target = sizes[-1] - 1

df_linear = test_linear_search(sizes, target)
df_bin = test_bin_search(sizes, target)

plt.plot(df_linear["n"], df_linear["time"], label="Linear Search")
plt.plot(df_bin["n"], df_bin["time"], label="Binary Search")
plt.xlabel("Розмір масиву (n)")
plt.ylabel("Час виконання (секунди)")
plt.title("Порівняння часу пошуку")
plt.legend()
plt.grid(True)
plt.show()
```

5. Оцінити, який з двох алгоритмів є ефективнішим і в якому діапазоні розміру задачі.

Лінійний пошук має складність 
𝑂(𝑛).
Бінарний пошук має складність 
𝑂(log⁡𝑛).
Отже, бінарний пошук ефективніший для великих масивів.
Лінійний пошук може бути прийнятним лише для дуже малих масивів або невідсортованих даних.
